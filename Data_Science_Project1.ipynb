##Background##

The objective of this project is to predict countries that produce rice but are at risk of low production and not able to meet consumption as a result of certain factors especially those related to climate change.  

Food insecurity is a growing global challenge and it is being exarcebated by the consequences of climate change which are rising temperatures, drought, irregular rainfall patterns etc. Most agrarian societies depend on the weather for cultivation of the land for food production.

Rice production is one of those activities where naturally rice is grown in swamp lands that have a natural supply of water coming from rainfall. 

Unfortunately, rainfall patterns are changing, and majority of these places used for growing rice get dried for the most part of the year. Or they become logged and overflowed with water an cannot becultivated. This condition is likely to affect rice production hence contribute to food insecurity. 

Being able to predict which regions are more likely at risk of low rice production can provide basis for adopting climate resilient approaches to rice cultivation.

```python
bq= ut.BigQuery()
qry =f"""
select distinct
  Area as name,
  Year as year,
  sum(case when element = "Yield" then Value else null end) as yield_hgha,
  sum(case when element = "Production" then Value else null end) as production_ton,
  sum(case when element = "Area harvested" then value else null end) as area_harvested_ha
from
  tarletondatascience2022.pometsey.rice
group by
  year,
  name
order by
  year,
  name
"""
rice_df = bq.qry_to_df(qry)
rice_df
#Data on rice yield
```

```python
countries='bigquery-public-data.ghcn_d.ghcnd_countries'
def make_qry(year):
    qry=f"""
    select distinct
        name,
        year,
        (tmax - tmin) as temp_diff,
        prcp
    from(
        select 
                id,
                name,
                element, 
                value,
                year,
                max(case when element = "TMAX" then value else null end) over (partition by year, name) as tmax,
                min(case when element = "TMIN" then value else null end) over (partition by year, name) as tmin,
                sum(case when element = "PRCP" then value else null end) over (partition by year, name) as prcp
        from(
            select
                    id,
                    element,
                    value,
                    date,
                    qflag,
                    {year} as year,
                    left(id,2) as country_code
            from
                bigquery-public-data.ghcn_d.ghcnd_{year} 
            where
                qflag != 'None' and
                element in ('TMAX','TMIN','PRCP') and
                value != 0.0
            )
        join
            bigquery-public-data.ghcn_d.ghcnd_countries as B
        on
            country_code= B.code
        )
    """ 
    return qry  


queries = [make_qry(year) for year in range(1961,2022)]
qry= 'union all'.join(queries) +'order by name, year asc'
w_df=bq.qry_to_df(qry)
w_df
#Data on temperature and precipitation
```

```python
print(w_df.dtypes)   #show data types

#convert numerical columns to float
w_df['temp_diff'] = w_df['temp_diff'].astype(float)
w_df['prcp'] = w_df['prcp'].astype(float)

# repalce nas with country median

w_df[['temp_diff','prcp']] = w_df.groupby(['name'])['temp_diff','prcp']\
    .transform(lambda x: x.fillna(x.median()))                        
print(w_df)
```

#Applying the Model

```python
#Standardized the data
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
df = final_df.drop('name', axis=1)       #remove name(string variable) from the data
scaler = StandardScaler()
df_scaled=scaler.fit_transform(df)
df_scaled= pd.DataFrame(df_scaled, columns=['year','temp_diff','prcp','ch4_emissions_kt','yield_hgha','production_ton','area_harvested_ha','population'])
df_scaled.head(5)
```

```python
#Do a pair plot to size the shape of the relationship between the variables
sns.pairplot(df_scaled)
```

Some clusters are visible but not very visible accross all the variables. Moreover, the cluster shapes are irregular.

```python
#much of the plot does not show any regular shape so we apply DBSCAN
#to selected column
#Two parameters need to be determined, eps and min_samples. The literature recommends we select min samples
# as 2*dimensions of data. There are various recommnedations for eps but I used the average nearest neighbor distance
# suggested to be calculated using 'kneelocator' from the 'kneed' package
from sklearn.neighbors import NearestNeighbors
X= df_scaled[['year','temp_diff','prcp','ch4_emissions_kt','yield_hgha','production_ton','area_harvested_ha','population']]

# n_neighbors = 5 as kneighbors function returns distance of point to itself (i.e. first column will be zeros) 
nbrs = NearestNeighbors(n_neighbors = 5).fit(X)
# Find the k-neighbors of a point
neigh_dist, neigh_ind = nbrs.kneighbors(X)
# sort the neighbor distances (lengths to points) in ascending order
# axis = 0 represents sort along first axis i.e. sort along row
sort_neigh_dist = np.sort(neigh_dist, axis = 0)
k_dist = sort_neigh_dist[:, 4]
plt.plot(k_dist)
plt.ylabel("k-NN distance")
plt.xlabel("Sorted observations (4th NN)")
plt.show()
```

```python
! pip install kneed              #install kneed
from kneed import KneeLocator    # import KneeLocator from kneed
kneedle = KneeLocator(x = range(1, len(neigh_dist)+1), y = k_dist, S = 1.0, 
                      curve = "concave", direction = "increasing", online=True)    #compute the 

# get the estimate of knee point
print('The estimated eps value is', kneedle.knee_y, 'and the min_sample value is ', 2*8)
```

```python
#we fit DBSCAN using the above best hyperparameter values
from sklearn.cluster import DBSCAN
dbs = DBSCAN(eps=3.7, min_samples=16)
clusters = dbs.fit(X)
X['labels'] = clusters.labels_.astype('str')
X.head(10)
```

```python
#we use KNeighbors using to compute a score
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score

Xtrain = X.drop('labels', axis=1)
ypred = X['labels']
knn=KNeighborsClassifier()
parameters = {
 'n_neighbors':[2,3,4,5],
 'algorithm':("auto", "ball_tree", "kd_tree", "brute")
}
grdmodel = GridSearchCV(knn,parameters, cv=10, scoring = 'f1_micro')
grdmodel.fit(Xtrain, ypred)
print(pd.DataFrame(grdmodel.cv_results_))
print(grdmodel.best_estimator_)
print(grdmodel.best_score_)
```
